# Generating some data

Now, let's generate some data! First, we can generate some data points for a 'scatter plot'. This is done by generating two lists, $x = [x_1, x_2, ...]$ and $y = [y_1, y_2, ...]$ so that we can plot the coordinate pairs $(x_1, y_1)$ and so on. In the code box below, define two lists and fill them with the following numbers:

x: 1.1, 3.2, 4.4, 5.2, 6.3  
y: 0.5, 3.5, 4.5, 4.8, 6.2

````python
#define your lists here
x = [1.1, 3.2, 4.4, 5.2, 6.3]
y = [0.5, 3.5, 4.5, 4.8, 6.2]
````

Now we are ready to plot our data! Since we are dealing with sets of coordinates, a scatter plot is a good choice for visualizing this data. For this, we will use the ```scatter()``` function from ```matplotlib.pyplot```.  

The ```scatter()``` function takes two inputs, which are your x and y coordinate lists. You can also add other inputs to customize your plot by changing the color, size, and shape of the scatter plot points.  

The final piece of the puzzle is actually making your plot appear. This is done using the ```show()``` function, which takes no inputs.  

In the code box below, make a scatter plot of your data by typing 

```
plt.scatter(x, y)
plt.show()
```
A few things to note:

*   We put ```plt.``` before the function to tell the code it must get the function from the ```matplotlib.pyplot library```
*   You must call the ```show()``` function after you generate the plot, since the code runs line by line. If you put the ```show()``` function before the ```scatter()``` function, it won't show anything because the plot hasn't been made yet!
*   If you want to know about the other options for the scatter function, the ```matplotlib.pyplot``` online documentation is very thorough. You can find it [here](https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.scatter.html).

````python
#generate your scatter plot and show it in this box
plt.scatter(x, y)
plt.show()
````

And just like that, we have made our first Python plot! Let's do a simple customization. Python has several color options for scatter plots. You can tell python which color you want using the 'color' parameter, and setting it equal to a string indicating the color you have chosen. Below is a list of all the 'named' colors available in matplotlib.pyplot:

named_colors.png



Select your favorite color and remake your plot, now including a color option in the ```scatter()``` function. For example:

```
plt.scatter(x, y, color = 'magenta')
plt.show()
```

````python
#make your plot with a custom point color in this code box
plt.scatter(x, y, color = 'magenta')
plt.show()
````

Finally, every good plot needs to have annotations. This means a title, axis labels, and perhaps a legend. This can all be achieved using matplotlib.pyplot functions, detailed below:

```
plt.scatter(x, y, color = 'magenta', label = 'My data')
plt.title('My title')
plt.xlabel('x value')
plt.ylabel('y data')
plt.legend(loc = 'upper left')
plt.show()
```
Remember to put ```plt.show()``` last, because any additional changes you make to your plot after the ```show()``` line will not show up!  

Note:
1.   ```title()```, ```xlabel()```, and ```ylabel()``` all take strings as their inputs
2.   Calling the ```legend()``` function with no inputs will have the program automatically select the best location for the legend. However, you can also specify the location by setting ```loc``` equal to a string such as ```'upper left'```, ```'lower right'```, and so on.
3.   The ```legend()``` function requires you to specify a label in the ```scatter()``` function. This is what will show up in the legend.

Plot your data once more in the box below, this time with a title, axis labels, and a legend. 

````python
#plot your data again in this text box
plt.scatter(x, y, color = 'magenta', label = 'My data')
plt.title('My title')
plt.xlabel('x value')
plt.ylabel('y value')
plt.legend(loc = 'upper left')
plt.show()
````

And there we have our first real Python plot! These techniques will be very useful for anything from school projects to future employment. Remember, there is a TON of documentation online for ```matplotlib.pyplot``` functions. If you have a question, chances are someone has already answered it online. Happy plotting!

# More Plotting

We can make more than just scatter plots! We can also make plots of mathematical functions. Let's try making a plot of $f(x) = \cos(x)$. For this, we will need to import ```numpy``` as ```np```, since this is the library that has the ```cos()``` function in it. Import ```numpy``` in the box below by typing  

```import numpy as np```

````python
#import numpy here as np
import numpy as np
````

Now that we have imported the library with our function, we want to make a plot of it. We will be using the function ```plot()``` from ```matplotlib.pyplot```. The main difference between ```plot()``` and ```scatter()``` is that ```plot()``` simply connects the coordinates you give it with a line, rather than placing markers at each point. You can see the documentation for ```plot()``` [here](https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.plot.html). 

Now we need to make the x coordinates that we will input into our function to get the y coordinates for our plot. To do this, we will use the ```linspace()``` function from ```numpy```, another common Python library.  

We can use the ```linspace()``` function to create an evenly spaced list of numbers. Let's say we want our x values on our plot to lie between -5 and 5. To make our plot look smooth, let's make a list of 100 points. We can make our list as follows:

```x = np.linspace(-5, 5, 100)```

What is this code doing?

We are defining a variable ```x```, which is equal to the output of our ```linspace()``` function, which is a list of 100 numbers evenly spaced between -5 and 5. You can try printing ```x``` in the box below to make sure it worked.

You can find documentation for ```linspace()``` [here](https://numpy.org/doc/stable/reference/generated/numpy.linspace.html).

````python
#make the x coordinates in this box and print them
x = np.linspace(-5, 5, 100)
print(x)
````

Finally, we can plot our function. To do this, we will use the ```plot()``` function from ```numpy```. In the box below, create a plot of $cos(x)$. Remember to use the ```show()``` function!

Hint: to apply $cos(x)$ to our list of x values, you will need to write 

```np.cos(x)```

You can find ```cos(x)``` documentation [here](https://numpy.org/doc/stable/reference/generated/numpy.cos.html).

````python
#plot cos(x) in this box
plt.plot(x, np.sin(x), label = 'sin(x)')
plt.plot(x, np.cos(x), label = 'cos(x)')
plt.plot(x, np.cos(x) + np.sin(x), label = 'cos(x) + sin(x)' )
plt.legend()
plt.show()
````
